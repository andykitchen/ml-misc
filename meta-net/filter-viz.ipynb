```python
import numpy as np

def plot_features(W, cols, h=28, pad=2):
    n_feats = W.shape[1]
    h = 28
    hh = 28 + 2*pad
    rows = n_feats / cols
    im = np.zeros((hh * rows, hh * cols))

    Ws = W[:, np.argsort(np.linalg.norm(W, axis=0))[::-1]]
    
    for i in xrange(n_feats / cols):
        for j in xrange(cols):
            im[i*hh+pad:(i+1)*hh-pad, j*hh+pad:(j+1)*hh-pad] = Ws[:,i*5 + j].reshape(28,28)

    plt.figure(figsize=(2*cols, 2*rows))
    plt.imshow(im, cmap=plt.gray(), interpolation='none');
```
